```python
import nibabel as nib
import numpy as np

import os
import glob
```

```python
# mount google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
root_dir = "/content/drive/MyDrive/ICH seg/ICH dataset"
```

```python
path_ncct_image = os.path.join(root_dir, "ct_scans")
output_dir = os.path.join(root_dir, "output")
```

```python
# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Find all NIfTI files in the 'ct_scans' subdirectory
nifti_files = glob.glob(os.path.join(path_ncct_image, "*.nii"))
```

```python
def skull_stripping(image, threshold):
    # Simple threshold-based skull stripping
    brain_mask = image > threshold
    brain = image * brain_mask
    return brain
```

```python
for i, path_ncct_image in enumerate(nifti_files):
    print(f'Processing {i+1}/{len(nifti_files)}: {path_ncct_image}')

    # Define the output path for the processed file
    path_ncct_brain = os.path.join(output_dir, os.path.basename(path_ncct_image).replace(".nii", "_brain.nii"))

    # Load the subject image
    img_subj_nii = nib.load(path_ncct_image)
    img_subj = img_subj_nii.get_fdata()
    img_subj_hdr = img_subj_nii.header

    # Skull stripping
    ncct_thr = 100  # Threshold for NCCT images
    brain = skull_stripping(img_subj, ncct_thr)

    # Save the brain image
    brain_img = nib.Nifti1Image(brain.astype(np.int16), img_subj_nii.affine, img_subj_hdr)
    nib.save(brain_img, path_ncct_brain)

    print(f'{path_ncct_image} ---- skull stripping finished and saved to {path_ncct_brain}')
```
